Objective:

The objective is to develop a web scraping solution that extracts specific data points in CSV format, enabling users to leverage the information for further analysis, insights, or reporting.

Tasks:

1.Identify Target Websites: Select and evaluate websites from which to scrape data based on the information required.

2.Set Up the Environment: Install necessary libraries and tools for web scraping (e.g., Beautiful Soup or requests in Python).

3.Develop the Web Scraper: Write scripts to navigate the target websites and extract specific data points.

4.Data Cleaning: Clean and format the scraped data to ensure consistency and usability.

5.Save Data in CSV Format: Store the cleaned data in a structured CSV file for easy access and analysis.

```python
# import libraries
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#URL web page get respose of side
r=requests.get("https://webscraper.io/test-sites/e-commerce/static/computers/tablets")
```

```python
r
```

```python
soup=BeautifulSoup(r.content,"html.parser")
```

```python
# Arange data structured formal
soup.prettify
```

```python
# web single page code
box=soup.find("div",class_="col-lg-9")

# create list of store data
product_name=[]
product_review=[]
product_price=[]
product_dis=[]
product_image=[]
```

```python
# Name of tablets/computers
name=box.find_all("a",class_="title")
for i in name:
    name=i.text
    product_name.append(name)
    
```

```python
len(product_name)
```

```python
# Price of tablets/computers
price=box.find_all("h4",class_="float-end price card-title pull-right")
for i in price:
    price=i.text
    product_price.append(price)
```

```python
len(product_price)
```

```python
# Review of tablets/computers
review=box.find_all("p",class_="float-end review-count")
for i in review:
    review=i.text
    product_review.append(review)
```

```python
len(product_review)
```

```python
# Discription of tablets/computers
dis=box.find_all("p",class_="description card-text")
for i in dis:
    dis=i.text
    product_dis.append(dis)
```

```python
len(product_dis)
```

```python
#Image link of Tablets/computers
image=box.find_all("img",class_="img-fluid card-img-top image img-responsive")
for i in image:
    image=i.get("src")
    product_image.append(image)
```

```python
len(product_image)
```

```python
#extract alternative of data 
alt=[]
image=box.find_all("img",class_="img-fluid card-img-top image img-responsive")
for i in image:
    image=i.get("alt")
    alt.append(image)
#     print(alt)
```

```python
len(alt)
```

```python
df=pd.DataFrame({"product_Name":product_name,"product_price":product_price,"product_Review":product_review,
                 "product_Discription":product_dis,"product_Image":product_image,"product_Alternative":alt})
df
        
```

```python
df.to_csv("Tablets_computer_data.csv")
```
